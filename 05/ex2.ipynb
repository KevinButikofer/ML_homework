```python
#Bütikofer Kevin, Charles-Lewis Jaggi
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt 

xl_file = pd.ExcelFile("overfitting.xlsx")

data_train =  np.array(xl_file.parse("train_set")) # sheet name
data_cv = np.array(xl_file.parse("cv_set")) # sheet name

x_train = data_train[:,0]
x_cv = data_cv[:,0]

n_train= x_train.shape[0]
n_cv= x_cv.shape[0]

y_train = data_train[:,1]
y_cv = data_cv[:,1]
```

```python
plt.scatter(x_train,y_train, color='black')
```

```python
def J_theta(h,y,n):
    return np.sum(np.square(h-y))/(2*n)
```

```python
new_x = np.linspace(-1, 11, 100)
cost_train=[]
cost_cv=[]
for i in range(1,8):
    thetas = np.polyfit(x_train, y_train, deg = i)
    
    h = np.polyval(thetas, new_x)
    h_train = np.polyval(thetas, x_train)
    h_cv = np.polyval(thetas, x_cv)
    
    plt.plot(new_x,h)
    cost_train.append(J_theta(h_train,y_train,n_train))
    cost_cv.append(J_theta(h_cv,y_cv,n_cv))
    
plt.scatter(x_train,y_train, color='black')
```

```python
plt.plot(cost_train, color = 'red')
plt.plot(cost_cv, color = 'blue')
```
