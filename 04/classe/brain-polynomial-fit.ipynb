```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
%matplotlib inline
```

```python
data = pd.read_csv('x02.csv', names=['gender', 'age', 'head_size', 'brain_weight'])
x = data['brain_weight'].values * 1.0
y = data['head_size'].values * 1.0
```

```python
plt.scatter(x,y)
```

```python
plt.hist(x, edgecolor='black')
```

```python
coef_deg1 = np.polyfit(x, y, deg=1)
coef_deg2 = np.polyfit(x, y, deg=2)
h1 = np.poly1d(coef_deg1)
h2 = np.poly1d(coef_deg2)
```

```python
plt.scatter(x,y)
new_x = np.linspace(1000, 1600, 200)
plt.plot(new_x, h1(new_x), color='red')
plt.plot(new_x, h2(new_x), color='green')
```
