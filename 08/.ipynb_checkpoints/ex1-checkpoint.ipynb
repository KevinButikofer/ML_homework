```python
X1_1 = X1[np.where(label1 == 0)]
X1_2 = X1[np.where(label1 == 1)]
X1_3 = X1[np.where(label1 == 2)]
plt.scatter(X1_1[:,0], X1_1[:,1], c='blue')
plt.scatter(X1_2[:,0], X1_2[:,1], c='red')
plt.scatter(X1_3[:,0], X1_3[:,1], c='green')
plt.show()
```

```python
def get_closest(data, centroid):
    closestCen = np.zeros(len(X1))
    for i in range(len(X1)):
        closestCen[i] = np.argmin(np.sqrt((X1[i,0] - centroid[:,0])**2 + (X1[i,1] - centroid[:,1])**2))
    return closestCen

def compute_centroids(data, , closestCen):
    newCentroid = np.zeros_like(centroid)
    for i in range(len(newCentroid)):
        d = X1[np.where(closestCen == i)]  
        if len(d) > 0:
            newCentroid[i] = [np.mean(d[:,0]), np.mean(d[:,1])]
    return newCentroid      
def converge(centroid, newCentroid):
    if (newCentroid==centroid).all():
        return [True, 0]
        
    dist = 0
    for c, n in zip (centroid, newCentroid):        
        dist += np.sqrt((n[0] - c[0])**2 + (n[1] - c[1])**2)        
    print(dist)
    
    return [True if dist == 0.0 else False, dist]
```

```python
def kMean(X1, K):
    centroid = X1[np.random.choice(len(X1), K)]
    closestCen = np.zeros(len(X1))

    finished = False
    while(not finished):
        get_closest(X1, centroid, closestCen)
        newCentroid = compute_centroids(X1, newCentroid, closestCen)
        conv = converge(centroid, newCentroid)
        finished = conv[0]
    return centroid, conv[1]
    
```

```python
res = kMean(X1, 3)
```
